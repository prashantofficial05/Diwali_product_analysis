```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df =pd.read_csv('Diwali sales product.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
#Exploratory Data Analsis
df.columns
```

```python
ax=sns.countplot(x = 'Gender',data=df)
```

```python
ax=sns.countplot(x = 'Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen =df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Gender', y='Amount',data = sales_gen)
```

```python
##from above graph we can see that most of the buyer are female and even th e purchasing power of females are greater than men 
```

```python
##Age 
```

```python
df.columns
```

```python
ax=sns.countplot(data=df, x = 'Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount and age group 
sales_age =df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x= 'Age Group', y='Amount',data = sales_age)
```

```python
#from above graph we can see that most  of the buyer are of age group between 26-35 yrs females
```

```python
#State 
```

```python
df.columns
```

```python
#total orders from states
sales_state =df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(13)


sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state,x ='State',y='Orders')
```

```python
#total amount/sales from top states
sales_state =df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(13)


sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state,x ='State',y='Amount')
```

```python
#marital Status 

ax =sns.countplot(data = df, x ='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(13)


sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state,x ='Marital_Status',y='Amount',hue='Gender')
```

```python
#From above graphs we can see that most of the buyer are married (women) and they have high purchasing power
```

```python
#Occupation 
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data =df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(13)


sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x ='Occupation',y='Amount')
```

```python
#from above grphas we can see that most buyer are wroking in IT Aviation and Healthcare 
```

```python
df.columns
```

```python
#Product Category 
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data =df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)


sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x ='Product_Category',y='Amount')
```

```python
#from above grphas we can see that most of the sold product are from food, footwear ,and Elctronics category
```

```python
sales_state =df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)


sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x ='Product_ID',y='Orders')
```

```python
#Conclusion
#Married women age group 26-35 yrs from up, Maharashtra and karnataka working in IT, Healthcare and Aviation are more likely to buy product from food clothing and Electronics category 
```
